```python
def player_salaries(team_name):

    known_salaries = []

    unknown_salaries = []

    url = f'https://www.spotrac.com/nhl/{team_name}/cap'

    spotrac = requests.get(url)

    soup = BeautifulSoup(spotrac.text, 'html.parser')

    base_url = 'https://www.spotrac.com/redirect'

    stats_links = []

    for link in soup.find_all('a'):

        url = link.get('href')

        if url and '/player/' in url:

            stats_links.append(url)

        stats_links = [item for item in stats_links if '/0/' not in item]

    for player_number in stats_links:

        player_url = player_number + 'cash-earnings/'

        player_request = requests.get(player_url)

        player_soup = BeautifulSoup(player_request.text, 'lxml')

        try:

            salaries = player_soup.body.find('div', class_='teams').find('ul', class_='cd-tabs-content')\
                        .li.section.table.get_text().split()
        except:

            unknown_salaries.append(player_number)

        player_name = [soup.h1.text]

        for money in salaries:

            player_name.append(money)
            
    return known_salaries
```
